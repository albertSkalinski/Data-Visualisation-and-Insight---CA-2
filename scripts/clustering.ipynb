# Clustering

## Importing Libraries

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

## Loading Data

```python
# Reading the user summary
userSummary = pd.read_csv(
    "C:/Users/sKALa/Repos/Semester 2/Data Visualisation and Insight/CAs/CA 2/Data-Visualisation-and-Insight---CA-2/data/preprocessed/userSummary.csv")
```

## Clustering

```python
# Performing k-means clustering
clusterColumns = ["totalSteps", "totalDistance", "veryActiveMinutes", "lightlyActiveMinutes", "sedentaryMinutes", "calories"]

X = userSummary[clusterColumns]

scaler = StandardScaler()
XScaled = scaler.fit_transform(X)

kMeans = KMeans(n_clusters = 3, random_state = 42, n_init = 10)
userSummary["cluster"] = kMeans.fit_predict(XScaled)
display(userSummary.head())
```

## Averaging

```python
# Displaying the mean of clusters
clusterSummary = userSummary.groupby("cluster")[clusterColumns].mean().reset_index()
display(clusterSummary)
```

## Plotting

```python
# Creating a plot
plt.figure()
sns.scatterplot(data = userSummary, x = "totalSteps", y = "calories", hue = "cluster", palette = "Set2", s = 100)

plt.title("User Clusters Based on Activity Behaviour")
plt.xlabel("Average Daily Steps")
plt.ylabel("Average Daily Calories")
plt.tight_layout()
plt.show()
```
